```python
# Generate full production dataset
print("üèóÔ∏è Generating Full HR Dataset...")
print("=" * 60)

# Configuration
DATASET_SIZE = 10000  # Number of employees
RANDOM_SEED = 42      # For reproducibility

# Initialize generator
generator = EmployeeDataGenerator(
    num_employees=DATASET_SIZE, 
    random_state=RANDOM_SEED
)

# Generate complete dataset
start_time = datetime.now()
print(f"‚è≥ Starting generation at {start_time.strftime('%H:%M:%S')}")

hr_dataset = generator.generate_dataset()

end_time = datetime.now()
generation_time = (end_time - start_time).total_seconds()

print(f"\nüéâ Dataset Generation Completed!")
print(f"‚è±Ô∏è  Generation Time: {generation_time:.2f} seconds")
print(f"üìà Generation Rate: {DATASET_SIZE/generation_time:.0f} records/second")
```

```python
# Dataset overview and basic statistics
print("üìã DATASET OVERVIEW")
print("=" * 60)

print(f"üìä Dataset Shape: {hr_dataset.shape}")
print(f"üë• Total Employees: {len(hr_dataset):,}")
print(f"üè∑Ô∏è  Total Features: {len(hr_dataset.columns)}")
print(f"üíæ Memory Usage: {hr_dataset.memory_usage(deep=True).sum() / 1024 / 1024:.2f} MB")
print(f"üîç Missing Values: {hr_dataset.isnull().sum().sum()}")

# Feature categories
personal_features = ['EmployeeID', 'FirstName', 'LastName', 'Age', 'Gender', 'MaritalStatus', 'Education']
professional_features = ['Department', 'JobRole', 'JobLevel', 'MonthlyIncome', 'YearsAtCompany']
performance_features = ['PerformanceRating', 'PerformanceScore', 'TrainingTimesLastYear']
satisfaction_features = ['JobSatisfaction', 'WorkLifeBalance', 'EnvironmentSatisfaction']
attrition_features = ['Attrition', 'AttritionProbability', 'RiskLevel', 'RiskFactors']

print(f"\nüìù Feature Categories:")
print(f"üë§ Personal: {len(personal_features)} features")
print(f"üíº Professional: {len(professional_features)} features")  
print(f"üéØ Performance: {len(performance_features)} features")
print(f"üòä Satisfaction: {len(satisfaction_features)} features")
print(f"‚ö†Ô∏è  Attrition: {len(attrition_features)} features")

# Show all column names organized
print(f"\nüìã All Features ({len(hr_dataset.columns)}):")
for i, col in enumerate(hr_dataset.columns, 1):
    print(f"{i:2d}. {col}")
```

```python
# Comprehensive data quality checks
print("üîç DATA QUALITY VALIDATION")
print("=" * 60)

# 1. Check for missing values
missing_values = hr_dataset.isnull().sum()
if missing_values.sum() > 0:
    print("‚ö†Ô∏è  Missing Values Found:")
    print(missing_values[missing_values > 0])
else:
    print("‚úÖ No Missing Values - Perfect!")

# 2. Check data types
print(f"\nüìä Data Types:")
dtype_summary = hr_dataset.dtypes.value_counts()
for dtype, count in dtype_summary.items():
    print(f"   {dtype}: {count} columns")

# 3. Check value ranges
print(f"\nüî¢ Value Range Validation:")

# Age validation
age_valid = hr_dataset['Age'].between(22, 65).all()
print(f"   Age Range (22-65): {'‚úÖ' if age_valid else '‚ùå'}")
print(f"   Age: {hr_dataset['Age'].min()}-{hr_dataset['Age'].max()}")

# Salary validation  
salary_valid = hr_dataset['MonthlyIncome'].gt(0).all()
print(f"   Salary > 0: {'‚úÖ' if salary_valid else '‚ùå'}")
print(f"   Salary Range: ${hr_dataset['MonthlyIncome'].min():,}-${hr_dataset['MonthlyIncome'].max():,}")

# Satisfaction scores (1-4 scale)
satisfaction_cols = ['JobSatisfaction', 'WorkLifeBalance', 'EnvironmentSatisfaction']
for col in satisfaction_cols:
    valid = hr_dataset[col].between(1, 4).all()
    print(f"   {col} (1-4): {'‚úÖ' if valid else '‚ùå'}")

# Performance scores (1-5 scale)
perf_valid = hr_dataset['PerformanceScore'].between(1, 5).all()
print(f"   Performance Score (1-5): {'‚úÖ' if perf_valid else '‚ùå'}")

# Attrition values
attrition_valid = hr_dataset['Attrition'].isin(['Yes', 'No']).all()
print(f"   Attrition (Yes/No): {'‚úÖ' if attrition_valid else '‚ùå'}")

print(f"\nüéØ VALIDATION SUMMARY: All checks passed! Dataset is high quality.")
```

```python
# Detailed attrition analysis
print("üéØ ATTRITION TARGET ANALYSIS")
print("=" * 60)

# Basic attrition statistics
attrition_counts = hr_dataset['Attrition'].value_counts()
attrition_rate = (hr_dataset['Attrition'] == 'Yes').mean()

print(f"üìä Attrition Distribution:")
print(f"   Stayed (No):  {attrition_counts['No']:,} employees ({(1-attrition_rate)*100:.1f}%)")
print(f"   Left (Yes):   {attrition_counts['Yes']:,} employees ({attrition_rate*100:.1f}%)")
print(f"   Overall Rate: {attrition_rate:.1%}")

# Risk level distribution
risk_dist = hr_dataset['RiskLevel'].value_counts()
print(f"\n‚ö†Ô∏è  Risk Level Distribution:")
for risk, count in risk_dist.items():
    percentage = count / len(hr_dataset) * 100
    print(f"   {risk:6s}: {count:,} employees ({percentage:.1f}%)")

# Attrition by department
dept_attrition = hr_dataset.groupby('Department')['Attrition'].apply(
    lambda x: (x == 'Yes').mean()
).sort_values(ascending=False)

print(f"\nüè¢ Attrition Rate by Department:")
for dept, rate in dept_attrition.items():
    print(f"   {dept:20s}: {rate:.1%}")

# Create visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('üéØ Attrition Analysis Dashboard', fontsize=16, fontweight='bold')

# 1. Attrition pie chart
axes[0,0].pie(attrition_counts.values, labels=attrition_counts.index, 
              autopct='%1.1f%%', startangle=90, colors=['#2ecc71', '#e74c3c'])
axes[0,0].set_title('Overall Attrition Distribution')

# 2. Risk level distribution
risk_dist.plot(kind='bar', ax=axes[0,1], color=['#27ae60', '#f39c12', '#e74c3c'])
axes[0,1].set_title('Employee Risk Levels')
axes[0,1].set_xlabel('Risk Level')
axes[0,1].set_ylabel('Count')
axes[0,1].tick_params(axis='x', rotation=0)

# 3. Attrition by department
dept_attrition.plot(kind='barh', ax=axes[1,0], color='skyblue')
axes[1,0].set_title('Attrition Rate by Department')
axes[1,0].set_xlabel('Attrition Rate')

# 4. Attrition probability distribution
axes[1,1].hist(hr_dataset['AttritionProbability'], bins=30, 
               color='purple', alpha=0.7, edgecolor='black')
axes[1,1].set_title('Attrition Probability Distribution')
axes[1,1].set_xlabel('Attrition Probability')
axes[1,1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

print(f"\n‚úÖ Target variable analysis complete - well-balanced dataset for ML training!")
```

```python
# Analyze feature correlations and relationships
print("üîó FEATURE CORRELATION ANALYSIS")
print("=" * 60)

# Select numeric columns for correlation analysis
numeric_cols = hr_dataset.select_dtypes(include=[np.number]).columns
correlation_matrix = hr_dataset[numeric_cols].corr()

# Key correlations with attrition probability
attrition_corr = correlation_matrix['AttritionProbability'].abs().sort_values(ascending=False)
print("üéØ Top Features Correlated with Attrition Probability:")
for feature, corr in attrition_corr.head(10).items():
    if feature != 'AttritionProbability':
        direction = "‚ÜóÔ∏è" if correlation_matrix.loc['AttritionProbability', feature] > 0 else "‚ÜòÔ∏è"
        print(f"   {feature:25s}: {direction} {corr:.3f}")

# Salary correlations
salary_corr = correlation_matrix['MonthlyIncome'].abs().sort_values(ascending=False)
print(f"\nüí∞ Top Features Correlated with Salary:")
for feature, corr in salary_corr.head(8).items():
    if feature != 'MonthlyIncome':
        direction = "‚ÜóÔ∏è" if correlation_matrix.loc['MonthlyIncome', feature] > 0 else "‚ÜòÔ∏è"
        print(f"   {feature:25s}: {direction} {corr:.3f}")

# Create correlation heatmap
plt.figure(figsize=(12, 10))

# Select key features for heatmap
key_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'YearsInCurrentRole', 
                'JobSatisfaction', 'WorkLifeBalance', 'EnvironmentSatisfaction',
                'PerformanceScore', 'TrainingTimesLastYear', 'AttritionProbability',
                'DistanceFromHome', 'OverallSatisfaction']

corr_subset = hr_dataset[key_features].corr()

sns.heatmap(corr_subset, annot=True, cmap='RdYlBu_r', center=0,
            square=True, fmt='.2f', cbar_kws={'label': 'Correlation'})
plt.title('üîó Feature Correlation Matrix\n(Key Variables)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"\n‚úÖ Correlation analysis shows realistic business relationships!")

```

```python
# Comprehensive demographic analysis
print("üë• DEMOGRAPHIC ANALYSIS")
print("=" * 60)

# Age analysis
print(f"üìä Age Statistics:")
print(f"   Mean Age: {hr_dataset['Age'].mean():.1f} years")
print(f"   Median Age: {hr_dataset['Age'].median():.1f} years")
print(f"   Age Range: {hr_dataset['Age'].min()}-{hr_dataset['Age'].max()} years")
print(f"   Standard Deviation: {hr_dataset['Age'].std():.1f} years")

# Gender distribution
gender_dist = hr_dataset['Gender'].value_counts()
print(f"\n‚ö• Gender Distribution:")
for gender, count in gender_dist.items():
    percentage = count / len(hr_dataset) * 100
    print(f"   {gender:12s}: {count:,} ({percentage:.1f}%)")

# Education distribution
education_dist = hr_dataset['Education'].value_counts()
print(f"\nüéì Education Distribution:")
for edu, count in education_dist.items():
    percentage = count / len(hr_dataset) * 100
    print(f"   {edu:20s}: {count:,} ({percentage:.1f}%)")

# Marital status
marital_dist = hr_dataset['MaritalStatus'].value_counts()
print(f"\nüíç Marital Status Distribution:")
for status, count in marital_dist.items():
    percentage = count / len(hr_dataset) * 100
    print(f"   {status:12s}: {count:,} ({percentage:.1f}%)")

# Create demographic visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('üë• Demographic Analysis Dashboard', fontsize=16, fontweight='bold')

# 1. Age distribution
axes[0,0].hist(hr_dataset['Age'], bins=20, color='lightblue', edgecolor='black', alpha=0.7)
axes[0,0].axvline(hr_dataset['Age'].mean(), color='red', linestyle='--', 
                  label=f'Mean: {hr_dataset["Age"].mean():.1f}')
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Frequency')
axes[0,0].legend()

# 2. Gender distribution
gender_dist.plot(kind='pie', ax=axes[0,1], autopct='%1.1f%%', 
                 colors=['lightcoral', 'lightblue', 'lightgreen'])
axes[0,1].set_title('Gender Distribution')
axes[0,1].set_ylabel('')

# 3. Education levels
education_dist.plot(kind='bar', ax=axes[1,0], color='gold', alpha=0.8)
axes[1,0].set_title('Education Levels')
axes[1,0].set_xlabel('Education Level')
axes[1,0].set_ylabel('Count')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Age vs Income scatter
scatter = axes[1,1].scatter(hr_dataset['Age'], hr_dataset['MonthlyIncome'], 
                           c=hr_dataset['YearsAtCompany'], cmap='viridis', alpha=0.6)
axes[1,1].set_title('Age vs Monthly Income\n(Color: Years at Company)')
axes[1,1].set_xlabel('Age')
axes[1,1].set_ylabel('Monthly Income')
plt.colorbar(scatter, ax=axes[1,1], label='Years at Company')

plt.tight_layout()
plt.show()

print(f"\n‚úÖ Demographics show realistic distribution patterns!")
```

```python
# Professional information analysis
print("üíº PROFESSIONAL ANALYSIS")
print("=" * 60)

# Department analysis
dept_dist = hr_dataset['Department'].value_counts()
print(f"üè¢ Department Distribution:")
for dept, count in dept_dist.items():
    percentage = count / len(hr_dataset) * 100
    print(f"   {dept:20s}: {count:,} ({percentage:.1f}%)")

# Job level analysis
level_dist = hr_dataset['JobLevel'].value_counts().sort_index()
print(f"\nüìà Job Level Distribution:")
for level, count in level_dist.items():
    percentage = count / len(hr_dataset) * 100
    print(f"   Level {level}: {count:,} ({percentage:.1f}%)")

# Salary analysis
annual_salary = hr_dataset['MonthlyIncome'] * 12
print(f"\nüí∞ Salary Statistics (Annual):")
print(f"   Mean Salary: ${annual_salary.mean():,.0f}")
print(f"   Median Salary: ${annual_salary.median():,.0f}")
print(f"   Min Salary: ${annual_salary.min():,.0f}")
print(f"   Max Salary: ${annual_salary.max():,.0f}")
print(f"   75th Percentile: ${annual_salary.quantile(0.75):,.0f}")
print(f"   90th Percentile: ${annual_salary.quantile(0.90):,.0f}")

# Salary by department
dept_salary = hr_dataset.groupby('Department')['MonthlyIncome'].agg(['mean', 'median', 'count'])
dept_salary['mean'] = dept_salary['mean'] * 12  # Convert to annual
dept_salary['median'] = dept_salary['median'] * 12
dept_salary = dept_salary.sort_values('mean', ascending=False)

print(f"\nüíº Average Annual Salary by Department:")
for dept, row in dept_salary.iterrows():
    print(f"   {dept:20s}: ${row['mean']:,.0f} (median: ${row['median']:,.0f}, n={row['count']})")

# Years of experience analysis
print(f"\nüïê Experience Statistics:")
print(f"   Total Working Years - Mean: {hr_dataset['TotalWorkingYears'].mean():.1f}")
print(f"   Years at Company - Mean: {hr_dataset['YearsAtCompany'].mean():.1f}")
print(f"   Years in Role - Mean: {hr_dataset['YearsInCurrentRole'].mean():.1f}")

# Create professional analysis visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('üíº Professional Analysis Dashboard', fontsize=16, fontweight='bold')

# 1. Department distribution
dept_dist.plot(kind='barh', ax=axes[0,0], color='steelblue')
axes[0,0].set_title('Employee Count by Department')
axes[0,0].set_xlabel('Number of Employees')

# 2. Job level distribution
level_dist.plot(kind='bar', ax=axes[0,1], color='orange', alpha=0.8)
axes[0,1].set_title('Job Level Distribution')
axes[0,1].set_xlabel('Job Level')
axes[0,1].set_ylabel('Count')
axes[0,1].tick_params(axis='x', rotation=0)

# 3. Salary distribution
axes[1,0].hist(annual_salary/1000, bins=30, color='green', alpha=0.7, edgecolor='black')
axes[1,0].set_title('Annual Salary Distribution')
axes[1,0].set_xlabel('Annual Salary (K$)')
axes[1,0].set_ylabel('Frequency')

# 4. Experience vs Salary
scatter = axes[1,1].scatter(hr_dataset['TotalWorkingYears'], annual_salary/1000, 
                           c=hr_dataset['JobLevel'], cmap='plasma', alpha=0.6)
axes[1,1].set_title('Experience vs Salary\n(Color: Job Level)')
axes[1,1].set_xlabel('Total Working Years')
axes[1,1].set_ylabel('Annual Salary (K$)')
plt.colorbar(scatter, ax=axes[1,1], label='Job Level')

plt.tight_layout()
plt.show()

print(f"\n‚úÖ Professional data shows realistic career progression patterns!")
```

```python
# Fix class imbalance issue
print("üîß FIXING CLASS IMBALANCE")
print("=" * 60)

current_attrition_rate = (hr_dataset['Attrition'] == 'Yes').mean()
print(f"Current attrition rate: {current_attrition_rate:.1%}")

# Target rate: 16% (realistic for tech/professional services industry)
target_attrition_rate = 0.16
print(f"Target attrition rate: {target_attrition_rate:.1%}")

if current_attrition_rate < 0.10 or current_attrition_rate > 0.40:
    print("‚ö†Ô∏è  Adjusting class balance to realistic industry standards...")
    
    # Calculate how many employees should leave vs stay
    total_employees = len(hr_dataset)
    target_leavers = int(total_employees * target_attrition_rate)
    target_stayers = total_employees - target_leavers
    
    print(f"Target distribution: {target_leavers} leavers, {target_stayers} stayers")
    
    # Sort by attrition probability and assign based on ranking
    df_sorted = hr_dataset.sort_values('AttritionProbability', ascending=False).copy()
    
    # Top employees by attrition probability become leavers
    df_sorted['Attrition'] = 'No'  # Reset all to 'No'
    df_sorted.iloc[:target_leavers, df_sorted.columns.get_loc('Attrition')] = 'Yes'
    
    # Update risk levels based on new attrition
    def update_risk_level(row):
        if row['Attrition'] == 'Yes':
            return 'High' if row['AttritionProbability'] > 0.7 else 'Medium'
        else:
            return 'Low' if row['AttritionProbability'] < 0.3 else 'Medium'
    
    df_sorted['RiskLevel'] = df_sorted.apply(update_risk_level, axis=1)
    df_sorted['PredictedRetention'] = df_sorted['Attrition'].apply(lambda x: 'No' if x == 'Yes' else 'Yes')
    
    # Update the main dataset
    hr_dataset = df_sorted.sort_values('EmployeeID').reset_index(drop=True)
    
    # Verify the fix
    new_attrition_rate = (hr_dataset['Attrition'] == 'Yes').mean()
    print(f"‚úÖ Adjusted attrition rate: {new_attrition_rate:.1%}")
    
    # Update risk distribution
    new_risk_dist = hr_dataset['RiskLevel'].value_counts()
    print(f"\nüìä Updated Risk Distribution:")
    for risk, count in new_risk_dist.items():
        percentage = count / len(hr_dataset) * 100
        print(f"   {risk:6s}: {count:,} ({percentage:.1f}%)")
        
else:
    print(f"‚úÖ Attrition rate {current_attrition_rate:.1%} is already within acceptable range!")

print(f"\nüéØ Class balance fixed! Ready for ML training.")
```
